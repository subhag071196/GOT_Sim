```python
import in_place
```

### I used in_place to remove 'JON: ' and 'Jon Snow:' from the text

```python
with in_place.InPlace('jon.txt') as file:
    for line in file:
        line = line.replace('JON: While we\'re preparing for attack, we need to make weapons from it.', '\n')
        file.write(line)
```

```python
cnt = 0
with open('jon.txt','r') as file:
    for line in file:
        cnt = cnt+1
        print(cnt)
        print(line)
```
